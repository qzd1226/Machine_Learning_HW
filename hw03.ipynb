```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import fbeta_score, make_scorer
from sklearn.metrics import accuracy_score
```

```python
trnX = pd.read_csv ('trnX-200d.csv',sep = '\t', header = None)
print(trnX)
```

```python
trnX.shape
```

```python
trnY = pd.read_csv ('trnY-200d.csv',header = None)
trnY.shape
```

```python
print(trnY)
```

# Grid Search

Grid search scope

```python
tuning_parameters = {'hidden_layer_sizes':[(4,),(8,),(16,),(32)],
                     'activation':('logistic','tanh','relu'),
                     'solver':('sgd','adam'),
                     'learning_rate_init':[1.0,0.1,0.01,0.001,0.0001],
                     'batch_size':[8,16,32]
                    }
```

```python
#grid_search = GridSearchCV(estimator=MLPClassifier(), param_grid=tuning_parameters,

clf = GridSearchCV(MLPClassifier(), tuning_parameters, cv=5)
clf.fit(trnX, trnY.values.ravel()) # X is tra  in samples and y is the corresponding labels
```

Get the best parameters

```python
print(clf.best_params_)
```

# Train the model with the Best Parameters

```python
optClf = MLPClassifier(activation='relu',batch_size = 16, hidden_layer_sizes=(32,), solver='sgd',learning_rate_init = 0.01).fit(trnX, trnY)
```

```python
devX = pd.read_csv ('devX-200d.csv',sep = '\t', header = None)
devY = pd.read_csv ('devY-200d.csv',sep = '\t', header = None)
```

```python
devX.shape
```

```python
devPredict = optClf.predict(devX)
devPredict = devPredict.reshape(-1,1)
devOptAcc = 1 - ((abs(devPredict - devY)).sum())/270
devOptAcc
```

# Train the Model with the Default Model Parameters

```python
clfDefault = MLPClassifier().fit(trnX, trnY)
```

```python
defaultPredict = clfDefault.predict(devX)
defaultPredict = defaultPredict.reshape(-1,1)
defaultAcc =  1 - ((abs(defaultPredict - devY)).sum())/270
defaultAcc
```

The difference between the best accuracy and the accuracy produced by the default hyperparameters:

```python
devOptAcc - defaultAcc
```

Accuracy improvement:About 5%

```python
(devOptAcc - defaultAcc)/defaultAcc
```
